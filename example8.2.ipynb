```python
#same as last example but using implicit scheme

import matplotlib.pyplot as plt
import numpy
%matplotlib inline
```

```python
def diffusion(nx,tf,dt,x):
    
    #given information about the rod
    
    rhoc=10*10**6    
    k=10
    L=.05

    dx=L/nx
    
    nt=tf/dt
    
    lambdax=(k*dt)/(rhoc*dx**2)
    
    
    T=numpy.zeros(nx) 
    T[:]=200
    TN=T.copy() #Temp distrubtion at time =0


    for i in range(int(nt)):
        TN=T.copy()

        T[1:-1]= lambdax * (T[2:] - 2 * T[1:-1] + T[0:-2] ) + TN[1:-1]

        #boundary conditions

        T[0] = lambdax * ( T[1] - T[0] ) + TN[0] 
        
        T[-1] = lambdax * (-3 * T[-1] + T[-2] ) + TN[-1]
        
        

    
    if x==1:
        print(T)
    X=numpy.linspace(0,.02,nx)
    plt.title(f"timestep {dt} s")
    plt.xlabel("Distance (m)")
    plt.ylabel("Temeperature (C)")
    plt.plot(X,T,label=dx)

plt.figure(figsize=(11,11),dpi=100)
diffusion(5,40,2,1)
diffusion(5,80,2,1)
diffusion(5,120,2,1)
legend=plt.legend(["40 s ","80 s ","120 s "])
legend.set_title("Time")
```
